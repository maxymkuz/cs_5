```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stay = pd.read_csv("files/stay.csv", sep=';')
walk = pd.read_csv("files/walk.csv", sep=';')
run = pd.read_csv("files/run.csv", sep=';')
```

```python
cols = ['ACCELEROMETER X (m/s²)', 'ACCELEROMETER Y (m/s²)', 'ACCELEROMETER Z (m/s²)',
        'GYROSCOPE X (rad/s)', 'GYROSCOPE Y (rad/s)', 'GYROSCOPE Z (rad/s)', 'Time since start in ms ']
```

```python
def prepare(df):
    df = df[cols]
    df = df.rename(columns={'ACCELEROMETER X (m/s²)': 'a_x', 'ACCELEROMETER Y (m/s²)': 'a_y', 'ACCELEROMETER Z (m/s²)': 'a_z',
        'GYROSCOPE X (rad/s)': 'z_x', 'GYROSCOPE Y (rad/s)': 'z_y', 'GYROSCOPE Z (rad/s)': 'z_z', 'Time since start in ms ': 'time'})
    df['a_sum'] = (df['a_x'] ** 2 + df['a_y'] ** 2 + df['a_z'] ** 2) ** (1/2)
    df['z_sum'] = (df['z_x'] ** 2 + df['z_y'] ** 2 + df['z_z'] ** 2) ** (1/2)
    return df
    
    
```

```python
stay = prepare(stay)
walk = prepare(walk)
run = prepare(run)
```

## Accelerometer:

```python
time = run['time'] / 10 ** 3
x = run['a_x']
y = run['a_y']
z = run['a_z']

plt.title('Acceleration on 3 axis, running:')
plt.subplot(3, 1, 1)
plt.plot(time, x)
plt.xlabel('time (s)')
plt.ylabel('X axis')

plt.subplot(3, 1, 2)
plt.plot(time, y)
plt.xlabel('time (s)')
plt.ylabel('Y axis')

plt.subplot(3, 1, 3)
plt.plot(time, z)
plt.xlabel('time (s)')
plt.ylabel('Z axis')
```

```python
x = stay['time'] / 10 ** 3
y = stay['a_sum']

plt.subplot(3, 1, 1)
plt.plot(x, y, '.-')
plt.title('Total acceleration of 3 axis')
plt.ylabel('staying')

x = walk['time'] / 10 ** 3
y = walk['a_sum']
plt.subplot(3, 1, 2)
plt.plot(x, y, '.-')
plt.xlabel('time (s)')
plt.ylabel('walking')

x = run['time'] / 10 ** 3
y = run['a_sum']
plt.subplot(3, 1, 3)
plt.plot(x, y, '.-')
plt.xlabel('time (s)')
plt.ylabel('running')

plt.show()
```

## Gyroscope:

```python
x = stay['time'] / 10 ** 3
y = stay['z_sum']

plt.subplot(3, 1, 1)
plt.plot(x, y, '.-')
plt.title('Total gyroscope signal of 3 axis')
plt.ylabel('staying')

x = walk['time'] / 10 ** 3
y = walk['z_sum']
plt.subplot(3, 1, 2)
plt.plot(x, y, '.-')
plt.xlabel('time (s)')
plt.ylabel('walking')

x = run['time'] / 10 ** 3
y = run['z_sum']
plt.subplot(3, 1, 3)
plt.plot(x, y, '.-')
plt.xlabel('time (s)')
plt.ylabel('running')

plt.show()
```
